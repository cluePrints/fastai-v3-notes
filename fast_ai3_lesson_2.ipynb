```
from torch import tensor
from torch import nn
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```
n = 100
x = torch.ones(n)
x.uniform_(0, 1)
x[:5]
```

```
orig_params = tensor([3.,1])
```

```
# Note to self: 2nd column was needed to vectorize ax+b: a*x+b*1 = [a,b] . [x, 1]
x__matrix = torch.cat([x.unsqueeze(1), torch.ones(n).unsqueeze(1)], dim=1)
x__matrix[:5]
```

```
y = x__matrix@orig_params + torch.rand(n)
```

```
y[:5]
```

```
p = nn.Parameter(tensor([1.,0]))
```

```
def mse_loss(y, y_hat):
  return ((y-y_hat)**2).mean()
```

```
(x__matrix@p)[:5], y[:5]
```

```
mse_loss(y, x__matrix@p)
```

```
 # Note to self: otherwise animation.to_jshtml() needed
from matplotlib import rc
rc('animation', html='jshtml')
```

```
lr = 0.1
def update(params, print_loss=False):
  y_hat = x__matrix@params
  loss = mse_loss(y, y_hat)
  if print_loss: print("Loss: " + str(loss))
  loss.backward()
  with torch.no_grad():
    p.sub_(p.grad * lr)
    # zeroing grad is to prevent accumulation, otherwise it will grow we won't converge
    p.grad.zero_()
```

```
a = tensor([1.,0])
p = nn.Parameter(a)

from matplotlib.animation import FuncAnimation
fig = plt.figure()
line, = plt.plot(x.numpy(), (x__matrix@a).numpy())
plt.scatter(x, y, c='orange')
plt.close()

def animate(i):
  line.set_ydata(x__matrix@a)
  print_loss = (i % 10 == 0)
  update(p, print_loss)
 
FuncAnimation(fig, func=animate, frames=np.arange(0,100), interval=20)
```
