```
def annealer(func):
  def _result(min_lr, max_lr): return partial(func, min_lr, max_lr)
  return _result

@annealer
def linear(min_lr, max_lr, progress):
  return min_lr + (max_lr - min_lr) * progress

test_eq(0.2,  linear(0.2, 0.3) (0))
test_eq(0.25, linear(0.2, 0.3) (0.5))
test_eq(0.3,  linear(0.2, 0.3) (1))
```

```
@annealer
def cos(min_lr, max_lr, progress):
  mult = (1 + math.cos(math.pi*(1-progress)))/2
  return min_lr + (max_lr - min_lr) * mult

test_eq(0.2,   cos(0.2, 0.3) (0))
test_eq(0.25,  cos(0.2, 0.3) (1/2))
test_eq(0.3,   cos(0.2, 0.3) (1))
```

```
@annealer
def no(min_lr, max_lr, progress):
  return min_lr

test_eq(0.2, no(0.2, 0.3) (1))
test_eq(0.2, no(0.2, 0.3) (0))
test_eq(0.2, no(0.2, 0.3) (1/2))
```

```
@annealer
def exp(min_lr, max_lr, progress):
  ratio = max_lr / min_lr
  return min_lr * (ratio) ** progress

test_eq(1, exp(1, 4) (0))
test_eq(2, exp(1, 4) (1/2))
test_eq(4, exp(1, 4) (1))
```

```
def combine_scheds(pcts, scheds):
  assert sum(pcts) == 1
  assert len(pcts) == len(scheds)
  pcts = tensor(pcts)
  assert len(pcts.shape) == 1
  assert (pcts > 0).all()
  right_pcts_boundaries = pcts.cumsum(dim = 0)
  def _combined_sched(combined_progress):
    current_sched_idx = (right_pcts_boundaries >= combined_progress).nonzero().min()
    local_progress_left = right_pcts_boundaries[current_sched_idx] - combined_progress
    local_progress_pct = (pcts[current_sched_idx] - local_progress_left)/pcts[current_sched_idx]
    return scheds[current_sched_idx](local_progress_pct)
  
  return _combined_sched
      
combined = combine_scheds([0.2, 0.8], [linear(0.2, 0.6), linear(0.6, 0.3)])
test_eq(0.2, combined(0))
test_eq(0.6, combined(0.2))
test_eq(0.3, combined(1))
```

```
def plot_scheduler(sched):
  progress_vals = np.arange(0, 1, 1e-2)
  series = [sched(progress_val) for progress_val in progress_vals]
  plt.plot(series)

combined = combine_scheds([0.2, 0.8], [linear(0.2, 0.6), linear(0.6, 0.3)])
plot_scheduler(combined)
```

```
class Recorder(Callback):
  def begin_fit(self):
    self.metrics = defaultdict(list)

  def after_batch(self):
    if not self.in_train:
      return

    for idx, param_groups in enumerate(self.opt.param_groups):
      self.metrics[f'lr_{idx}'].append(param_groups['lr'])
    
  def after_loss(self):
    self.metrics['loss'].append(self.loss.item())
    
  def plot_lrs(self):
    for name in filter(lambda key: 'lr_' in key, self.metrics.keys()):
      plt.plot(self.metrics[name])

  def plot_loss(self):
    plt.plot(self.metrics['loss'])
```

```
# 'inlined' from get_model()
n_inputs = data.train_ds.x.shape[1]
model = nn.Sequential(nn.Linear(n_inputs, n_hidden), nn.ReLU(), nn.Linear(n_hidden, data.c))

weight_params = []
bias_params = []
for layer in model.children():
  if isinstance(layer, nn.Linear):
    weight_params.append(layer.weight)
    bias_params.append(layer.bias)
    
param_groups = [{'params': weight_params}, {'params': bias_params}]
opt = optim.SGD(param_groups, lr=0.8)
```

```
class DiscriminativeScheduler(Callback):
  def __init__(self, sched_funcs):
    self.sched_funcs = sched_funcs

  def begin_batch(self):
    assert len(self.sched_funcs) == len(self.opt.param_groups)
    progress = self.n_epochs / self.epochs
    for param, sched_func in zip(self.opt.param_groups, self.sched_funcs):
      param['lr'] = sched_func(progress)
```

```
weights_schedule = combine_scheds([0.3, 0.7], [cos(0.3, 1), cos(1, 0.2)])
biases_schedule = combine_scheds([0.3, 0.7], [cos(0.1, 0.6), cos(0.6, 0.1)])

learner = Learner(model, opt, loss_func, data)
recorder = Recorder()
run = Runner([AvgStatsCallback([accuracy]),
              DiscriminativeScheduler([weights_schedule, biases_schedule]),
             recorder])
run.fit(3, learner)

recorder.plot_lrs()
```
